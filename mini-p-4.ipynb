```python
def square(width,height):
    for i in range(height):
        print("|"+"--"+"|")

def triangle(height):
    for i in range(1,height+1):
        spaces = " "*(height-i)
        print(spaces + "/" + "-" * (2 * i - 2) + "\\")

def main():
    print("Square:")
    square(5, 3)

    print("\nTriangle:")
    triangle(4)

if __name__ == "__main__":
    main()
```
